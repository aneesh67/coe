```python
import pandas as pd
df=pd.read_excel("height_weight_data.xlsx")
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(range(len(df)),df['Height (cm)'],color='green',label='data points')
plt.xlabel('length')
plt.ylabel('height')
```

```python
correlation = df.corr()
print("Correlation Matrix:")
print(correlation)
```

```python
import seaborn as sns
# Visualizing the correlation with a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Height and Weight')
plt.show()

# Scatter plot to visualize the relationship
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Height (cm)', y='Weight (kg)', data=df)
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
sns.pairplot(df)
plt.suptitle("Pair Plot of Height and Weight", y=1.02)
plt.show()
```

```python
sns.pairplot(data=df,hue='Weight (kg)')
plt.show()
```
